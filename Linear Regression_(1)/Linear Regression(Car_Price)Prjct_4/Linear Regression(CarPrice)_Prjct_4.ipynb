#### dropping unnecessary columns

```python
df=df.drop(columns=['car_ID','symboling','CarName'])
```

```python
df.info()
```

### converting float to int

```python
df['boreratio'] = df['boreratio'].astype(int)
df['stroke'] = df['stroke'].astype(int)
df['compressionratio'] = df['compressionratio'].astype(int)
df['price'] = df['price'].astype(int)
df['wheelbase'] = df['wheelbase'].astype(int)
df['carlength'] = df['carlength'].astype(int)
df['carwidth'] = df['carwidth'].astype(int)
df['carheight'] = df['carheight'].astype(int)
```

```python
df.fuelsystem.value_counts()
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['fuelsystem'])
```

```python
df.enginetype.value_counts()
```

```python
df=df.drop(columns=['enginetype'])
```

```python
df.cylindernumber.value_counts()
```

```python
df=df.drop(columns=['cylindernumber'])
```

```python
df.enginelocation.value_counts()
```

```python
for i in range(0,205):
    if df['enginelocation'][i]=="front":
        df['enginelocation'][i]=1
    else:
        df['enginelocation'][i]=0
```

```python
df.head(10)
```

### transforming categorical to numerical

```python
for i in range(0,205):
    if df['drivewheel'][i]=="fwd":
        df['drivewheel'][i]=1
    elif (df['drivewheel']=="4wd").any():
        df['drivewheel'][i]=0
    else:
        df['drivewheel'][i]=2
```

```python
df['drivewheel'].nunique
```

```python
df=df.drop(columns=['carbody','aspiration'])
```

```python
df.doornumber.value_counts()
```

```python
for i in range(0,205):
    if df['doornumber'][i]=="four":
        df['doornumber'][i]=1
    else:
        df['doornumber'][i]=0
```

```python
df['doornumber']
```

```python
df.fueltype.value_counts()
```

```python
for i in range(0,205):
    if df['fueltype'][i]=="gas":
        df['fueltype'][i]=1
    else:
        df['fueltype'][i]=0
```

```python
df['fueltype']
```

```python
df.head(10)
```

```python
x=df.iloc[:,0:-1]
```

```python
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

#### train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_test
```

### Scaling


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
```

```python
x_test_scaled=scaler.fit_transform(x_test)
```

```python
x_test_scaled
```

### Applying Linear Regression because data has linear behaviour

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train_scaled,y_train)
```

```python
y_pred=lr.predict(x_test_scaled)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

### Got 85.5% accuracy
